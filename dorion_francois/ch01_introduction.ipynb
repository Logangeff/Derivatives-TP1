In the book, $f(t,T)$ and $F(t,T)$ respectively denote the forward and futures **prices**, at time $T$, for maturity $T$. Shreve (e.g.) uses $F(t) = S(t) - f(0,T)$ for the value at time $T$ of forward **contract** that was struck at the time-0 forward price for maturity $T$. We do not have a notation for such a contract...
